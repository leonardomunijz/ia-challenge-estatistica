```python
# Importar a biblioteca pandas
import pandas as pd
```

```python
# Versão do pandas
pd.__version__
```

```python
# Importar a biblioteca matplotlib
import matplotlib.pyplot as plt
```

```python
# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
# Crie o DataFrame
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
```

```python
# Converta a coluna de datas para o tipo datetime
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref'])
```

```python
# Ordenar o DataFrame pela coluna 'data_ref'
df_faturamento = df_faturamento.sort_values(by='data_ref')
```

```python
# Adicione uma coluna para o mês e ano
df_faturamento['mes_ano'] = df_faturamento['data_ref'].dt.to_period('M')
```

```python
df_faturamento
```

```python
df_faturamento.info()
```

```python
# Calcule a média das vendas
media_vendas = df_faturamento['valor'].mean()
print(f'Média das vendas: R${media_vendas:,.2f}')
```

```python
# Crie o gráfico de barras verticais
plt.figure(figsize=(12, 6))
plt.bar(df_faturamento['mes_ano'].astype(str), df_faturamento['valor'], color='skyblue')
plt.title('Faturamento por Mês')
plt.xlabel('Mês de Referência')
plt.ylabel('Valor (R$)')
plt.xticks(rotation=45)
plt.grid(axis='y')
```

```python
# Exiba o gráfico de barras
plt.tight_layout()
plt.show()
```

```python
# Crie o gráfico de linhas
plt.figure(figsize=(12, 6))
plt.plot(df_faturamento['mes_ano'].astype(str), df_faturamento['valor'], marker='o', linestyle='-', color='b')
plt.title('Faturamento ao Longo do Tempo')
plt.xlabel('Mês de Referência')
plt.ylabel('Valor (R$)')
plt.xticks(rotation=45)
plt.grid(True)
```

```python
# Exiba o gráfico de linhas
plt.tight_layout()
plt.show()
```
